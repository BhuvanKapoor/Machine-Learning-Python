```python
df = pd.read_csv('./Data/test-scores.csv')
df
```

```python
x = df.math
y = df.cs
```

```python
def gradient_descent (x,y):
    slope = intercept = 0
    n = len(x)
    iterations = 1000
    learning_rate = 0.00001
    
    for i in range(iterations):
        y_pred = slope*x + intercept
        mse = (1/n) * sum([val**2 for val in (y-y_pred)])
        slope_deriv = -(2/n) * sum(x*(y-y_pred)) 
        intercept_deriv = -(2/n) * sum((y-y_pred)) 
        slope = slope - learning_rate * slope_deriv 
        intercept = intercept - learning_rate * intercept_deriv 
        # if (math.isclose(mse[i],mse[i-1],1e-20,0.0)):
        #     break
        print(f"slope {slope}, intercept {intercept}, mse {mse}, iteration {i}")
```

```python
gradient_descent(x,y)
```
